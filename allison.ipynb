```python
import numpy as np
from PIL import Image
```

```python
# 顯示圖片
def show_(arr):
    return Image.fromarray(np.uint8(arr))
```

```python
p1 = Image.open('bj6.png').resize((900, 900))
arr = np.array(p1)

p2 = Image.open('small.png')
p2 = p2.convert('RGBA')
arr2 = np.array(p2)

h, w, k = arr2.shape

# i, j 插入位置
i = 30
j = 460

# 需要去背的地方
flter = (arr2[:, :, 3:] != 0) * np.ones((h, w, 3)).astype(bool) 
arr[i:i+h, j:j+w,][flter] =  arr2[:, :, 0:3][flter]

show_(arr)
```

```python
p = Image.fromarray(arr)
p.save('allison.png')
```
